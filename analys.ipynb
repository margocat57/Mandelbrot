# Анализ оптимизаций алгоритма построения множества Мальдеброта

## Наивная реализация запуск на g++ (-O1)

````python
import pandas as pd

df = pd.read_csv("results.csv", header=None, sep=";")
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

filtered = df[
    (df.iloc[:,0] == "mandelbrot1") &
    (df.iloc[:,2] == "O1") &
    (df.iloc[:,1] == "g++")
]

sns.kdeplot(
    data=filtered,
    x=filtered.iloc[:,3],
    hue=filtered.iloc[:,1],
    fill=True,           
    alpha=0.3,           
    linewidth=2          
)

plt.xlabel("Время работы для 100 генераций")
plt.ylabel("Плотность")
plt.title("Распределение времени вычисления множества Мандельброта для наивной реализации (флаг -O1)")
plt.show()
````

````python
subset1_1_g = df[
    (df.iloc[:,0] == "mandelbrot1") &
    (df.iloc[:,2] == "O1") &
    (df.iloc[:,1] == "g++" )
]

subset1_1_g.describe()
````

````python
import numpy as np
from scipy.stats import norm

mean = np.mean(subset1_1_g.iloc[:,3])
std = np.std(subset1_1_g.iloc[:,3], ddof=1)
n = len(subset1_1_g.iloc[:,3])

alpha = 0.05 
z = norm.ppf(1 - alpha/2)

lower = mean - z * std / np.sqrt(n)
upper = mean + z * std / np.sqrt(n)

print("Доверительный интервал:")
print(lower, upper)
````

# Объяснение разницы работы реализаций

## G++ vs clang++ флаг -О2

Оба компилятора заменяют часть инструкций на SIMD варианиы, однако clang++ делает это лучше: например код 

for(int j = 0; j < 4; j++){x[j]= x0[j];}

clang++ заменил на SIMD вариант, а g++ - нет.

Также clang++ на данном уровне оптимизации развернул циклы

g++ :

<img src="godbolt_img/2_2.png" width="80%">

clang++:

<img src="godbolt_img/2_2с.png" width="80%">

## G++ -O2 vs g++ -О3

Intrinsic инструкции стали применяться реже, примеры на скриншотах. Например, стоит смотреть на строчки кода:

for(int j = 0; j < 4; j++){x2[j]= x[j] * x[j];}  (1)

for(int j = 0; j < 4; j++){x[j] = x2[j] - y2[j] + x0[j];} (2)

g++ -O3 :

<img src="godbolt_img/2_3.png" width="80%">

g++ -O2:

<img src="godbolt_img/2_3с.png" width="80%">

## G++ vs clang++ флаг -О1

Рассмотрим код: double X0 = ((- (double)SUPER_WIDTH / 2) * DX) * SCALE_X * params->zoom + params->center_x + BEAUTY_COEF * params->zoom;

clang использует vfmadd231sd — это инструкция, которая делает умножение и сложение за один шаг, а gcc отдельно использует vmulsd и vaddsd, причем один из операндов находится в памяти, из-за этого если данные находятся в L1 кэше,то vmuld + vaddsd будут выполняться за 15 тактов, а vfmadd231sd примерно 5 тактов. Если же данные находятся в более дальних кэшах или памяти, то задержка для версии, скомпилированной gcc, может составлять десятки или сотни тактов.
