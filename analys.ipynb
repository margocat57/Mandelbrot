## Алгоритм 1. Наивная реализация запуск на g++ (-O0)

```python
import pandas as pd
df = pd.read_csv("results/results_v1_O0.csv", header=None, sep=";")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

subset1_g = df[df.iloc[:, 1] == "g++"].iloc[:, 2]

sns.kdeplot(subset1_g, fill=True, color="blue")
plt.title("Плотность распределения для наивной реализации(g++ -O0)")
plt.show()
```

```python
subset1_g.describe()
```

```python
import numpy as np
from scipy.stats import norm


mean = np.mean(subset1_g)
std = np.std(subset1_g, ddof=1)
n = len(df.iloc[:, 2])

alpha = 0.05 
z = norm.ppf(1 - alpha/2)

lower = mean - z * std / np.sqrt(n)
upper = mean + z * std / np.sqrt(n)

print("Доверительный интервал:")
print(lower, upper)
```
